# Deep Learning for Images with PyTorch
TorchVision is a PyTorch image library that provides useful tools, including transformations for image pre-processing, pre-trained CNN models and labeled image datasets for trainiing and testing. While working with PyTorch on images, this library will be used to ease multiple processes.

### 1) Image Classification with CNNs
- **Binary Classification:** Two distinct classes, *Sigmoid* activation function
- **Multi-class Classification:** Multiple classes, *Softmax* activation function
